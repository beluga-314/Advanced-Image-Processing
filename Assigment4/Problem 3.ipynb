```python
Directory_path = '/content/drive/MyDrive/assignment4_data/object_detection/'
```

```python
yaml_path = Directory_path + 'data.yaml'
train_path = Directory_path + 'train/images'
val_path = Directory_path + 'val/images'
test_path = Directory_path + 'test/images'
```

```python
# Open the YAML file for editing
with open(yaml_path, 'r') as f:
    config = yaml.load(f, Loader=yaml.FullLoader)

# Modify the paths to the train, test, and validation directories
config['train'] = train_path
config['test'] = test_path
config['val'] = val_path

# Write the updated dictionary back to the YAML file
with open(yaml_path, 'w') as f:
    yaml.dump(config, f)
```
